# Data Cleaning

### 1. Remove Invalid Rows

```python
df.columns
```

```python
# HelpfulnessDenominator = Number of people who have reviewed
# HelpfulnessNumerator = Number of people who find the review helpful.

# HelpfulnessNumerator must be less than HelpfulnessDenominator
# If it false then it considered as invalid 
```

```python
df['HelpfulnessNumerator'] > df['HelpfulnessDenominator'] ##invalid rows
```

```python
df[df['HelpfulnessNumerator']> df['HelpfulnessDenominator']]
```

```python
df_valid= df[df['HelpfulnessNumerator']<= df['HelpfulnessDenominator']]
```

```python
df_valid.shape
```

### 2. Remove Duplictae Rows

```python
df.columns
```

```python
# No user can type a same review on same time for different product. If all these features are same then we considered it as a duplicate row.
```

```python
df_valid.duplicated(['UserId','ProfileName','Time','Text'])
```

```python
df_valid[df_valid.duplicated(['UserId','ProfileName','Time','Text'])]
```

```python
data= df_valid.drop_duplicates(subset=['UserId','ProfileName','Time','Text'])
```

### 5. Sentiment Analysis

```python
import textblob
```

```python
from textblob import TextBlob
```

```python
data['Summary'][0]
```

```python
TextBlob('Good Quality Dog Food').sentiment.polarity   # calculates the polarity score ranging from -1 to +1
```

```python
def calculate_polarity(summary):
    try:
        return TextBlob(summary).sentiment.polarity
    except AttributeError:
        return 0
```

```python
data['Polarity'] = data['Summary'].apply(calculate_polarity)
```

```python
data.head()
```

```python
data['Polarity']<0
```

```python
data_negative_feedback = data[data['Polarity']<0]
```

```python
data_positive_feedback = data[data['Polarity']>0]
```

```python
data_neutral_feedback = data[data['Polarity']==0]
```
